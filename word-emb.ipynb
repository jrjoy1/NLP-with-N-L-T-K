3. Word Embeddings

এখানেই Word Embedding আসে।

Idea:

প্রতিটি শব্দকে একটি dense vector হিসেবে represent করা।

উদাহরণ:

King  = [0.82, 0.44, 0.91]
Queen = [0.80, 0.47, 0.89]
Man   = [0.11, 0.22, 0.55]
Woman = [0.12, 0.24, 0.53]

খেয়াল করো:

King ≈ Queen
Man ≈ Woman

কারণ vector কাছাকাছি।

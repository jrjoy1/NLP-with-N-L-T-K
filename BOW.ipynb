```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus = [
    "I love NLP",
    "I love Python",
    "Python loves NLP"
]
```

```python
cv = CountVectorizer()
```

```python
X = cv.fit_transform(corpus)
```

```python
print(cv.get_feature_names_out())
```

```python
print(X.toarray())
```

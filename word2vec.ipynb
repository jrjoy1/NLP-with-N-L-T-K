```python
from gensim.models import Word2Vec
```

```python
sentences = [
    ["i", "love", "nlp"],
    ["i", "love", "python"],
    ["python", "and", "nlp"]
]
```

```python
model = Word2Vec(
    sentences,
    vector_size = 10,
    window=2,
    min_count=1
)
```

```python
print(model.wv["nlp"])
```

```python
print(model.wv.most_similar("python"))
```

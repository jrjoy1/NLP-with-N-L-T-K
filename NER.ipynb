```python
text = "The Eiffel Tower was build by Steve Jobs in Canada."
```

```python
import nltk
```

```python
token = nltk.word_tokenize(text)
```

```python
token
```

```python
tag_tokens = nltk.pos_tag(token)
```

```python
tag_tokens
```

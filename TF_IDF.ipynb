```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
corpus = [
    "I love NLP",
    "I love Python",
    "I love Machine Learning"
]
```

```python
tfidf = TfidfVectorizer()
```

```python
X = tfidf.fit_transform(corpus)
```

```python
print(tfidf.get_feature_names_out())
```

```python
print(X.toarray())
```
